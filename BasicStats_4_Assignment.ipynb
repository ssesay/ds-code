```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.patches as pltPatches

df_1996 = pd.read_csv('MERGED1996_97_PP.csv')
df_2011 = pd.read_csv('MERGED2011_12_PP.csv')
df_2017 = pd.read_csv('MERGED2017_18_PP.csv')
```

```python
df_2011_table = pd.DataFrame()
df_2017_table = pd.DataFrame()
```

```python
# Adds the column of the state abbreviation 
df_2011_table ['STABBR'] = df_2011['STABBR']
df_2017_table ['STABBR'] = df_2017['STABBR']
```

```python
# Adds the Admin rate to a data table 
df_2011_table ['ADM_RATE'] = df_2011['ADM_RATE']
df_2017_table ['ADM_RATE'] = df_2017['ADM_RATE']
```

```python
#Add SAT critical thinking mid-point percentile 
df_2011_table ['SATVRMID'] = df_2011['SATVRMID'] 
df_2017_table ['SATVRMID'] = df_2017['SATVRMID']
```

```python
#Add SAT math mid-point percentile 
df_2011_table ['SATMTMID'] = df_2011['SATMTMID'] 
df_2017_table ['SATMTMID'] = df_2017['SATMTMID']
```

```python
#Add SAT writing mid-point percentile 
df_2011_table ['SATWRMID'] = df_2011['SATWRMID'] 
df_2017_table ['SATWRMID'] = df_2017['SATWRMID']
```

```python
#Groups schools by state abbreviation 
df_2011_table.groupby('STABBR').mean()
df_2017_table.groupby('STABBR').mean()
```

```python
# Plot the Admission Rate for 2011
plt.plot(df_2011_table['ADM_RATE'], color='blue')

plt.show()
```

```python
# Plot the SAT Writing score versus SAT Math score  
plt.scatter(x=df_2011_table['SATVRMID'], y=df_2011_table['SATMTMID'], color='purple')
plt.title('2011 SAT SCORE: Correlating Math Aptitude to Critical Thinking')
plt.xlabel('Critical Thinking Score')
plt.ylabel('Mathematices Score')
plt.show()
```

```python
# Compares correlation between 2011 and 2017 
plt.scatter(x=df_2011_table['SATVRMID'], y=df_2011_table['SATMTMID'], color='blue')
plt.scatter(x=df_2017_table['SATVRMID'], y=df_2017_table['SATMTMID'], color='purple')

leg_red = pltPatches.Patch(color='blue', label='Year 2011')
leg_blue = pltPatches.Patch(color='purple', label='Year 2017')
plt.legend(handles=[leg_red],loc=1)
plt.legend(handles=[leg_blue],loc=2)

plt.title('2011 vs 2017 Correlation Comparison')
plt.xlabel('Critical Thinking Score')
plt.ylabel('Mathematices Score')

plt.show()
```

```python
#Histogram representation of data
plt.hist(df_2011_table['SATMTMID'], bins=75, color='red', alpha=.5)
plt.hist(df_2017_table['SATMTMID'], bins=75, color='purple', alpha=.5)

plt.title('2011 vs 2017 Math Score Frequency')
plt.xlabel('Mathematics Score')

plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.scatter(x=df_2011_table['ADM_RATE'], y=df_2011_table['SATMTMID'], color='red')
plt.xlabel('Admission Rate')

plt.subplot(1,2,2)
plt.scatter(x=df_2017_table['ADM_RATE'], y=df_2017_table['SATMTMID'], color='blue')
plt.xlabel('Admission Rate')

plt.title('2011 vs 2017  Critical Thinking and Admission Correlation')

plt.tight_layout()
plt.show()
```
