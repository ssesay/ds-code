```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
poisson = np.random.poisson(1, 100)
poisson.sort()
bernoulli = np.random.binomial (1, .5, 100)
gamma = np.random.gamma(5, 1, 100)
binomial = np.random.binomial (20, 0.5, 100)
normal = np.random.normal(5, 30, 100)
normal.sort()
chisquare = np.random.chisquare(1, 100)
```

```python
#Poisson Graph
plt.hist(poisson, bins=10, color='c')
plt.axvline(poisson.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(np.mean(poisson) + poisson.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(poisson.mean()-poisson.std(), color='b', linestyle='dashed', linewidth=2) 
plt.show()

#Bernoulli Graph 
plt.hist(bernoulli)
plt.axvline(bernoulli.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(np.mean(bernoulli) + bernoulli.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(bernoulli.mean()-bernoulli.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()

#Gamma Graph 
plt.hist(gamma)
plt.axvline(gamma.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(np.mean(gamma) + gamma.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(gamma.mean()-gamma.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()

#Binomial Graph
plt.hist(binomial)
plt.axvline(binomial.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(np.mean(binomial) + binomial.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(binomial.mean()-binomial.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()

#Normal Graph
plt.hist(normal)
plt.axvline(normal.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(np.mean(normal) + normal.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(normal.mean()-normal.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()

#ChiSquare 
plt.hist(chisquare)
plt.axvline(chisquare.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(np.mean(chisquare) + chisquare.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(chisquare.mean()-chisquare.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```
