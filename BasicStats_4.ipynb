```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
plt.plot([0,1,2,3,4])
plt.show()
```

```python
np.random.seed (1221)
```

```python
df = pd.DataFrame()

df['rand'] = np.random.rand(100)
df['rand_sq'] = df['rand']**2
df['rand_shift'] = df['rand'] + 2 
```

```python
df['counts_sq'] = df.index ** 2
df['counts_sqrt'] = np.sqrt(df.index)
```

```python
plt.plot(df['rand'])
plt.show()
```

```python
plt.plot(df['rand'], color='purple')
plt.ylim([-0.1,1.1])
plt.ylabel('Values')
plt.title('Random Series')
plt.show()
```

```python
plt.plot(df['rand'], color='purple')
plt.plot(df['rand_shift'], color='green')
plt.ylim(-0.1,3.1)
plt.ylabel('Values')
plt.title('Random Series')
plt.show()
```

```python
plt.scatter(x=df['rand'], y=df['rand_sq'])
plt.show
```

```python
plt.scatter(x=df['rand'],y=df['rand_sq'],color='purple',marker='x',s=10)
plt.scatter(x=df['rand'],y=df['rand_shift'],color='green',marker='x',s=10)
plt.show()
```

```python
print('Pandas Techniques for Printing Plots')
```

```python
df.plot(kind='scatter', x='counts_sq', y='counts_sqrt')
df.plot(kind='line')
plt.show()
```

```python
plt.figure (figsize=(10,5))

plt.subplot(1, 2, 1)
plt.plot(df['rand'], color='purple')
plt.ylabel('Values')
plt.title('Random Series')

plt.subplot(1, 2, 2)
plt.plot(df['rand_shift'], color='green')
plt.ylabel('Shifted Values')
plt.title('Shifted Series')
plt.show
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1, 2, 1)
plt.plot(df['rand'], color='purple')
plt.ylabel('Values')
plt.title('Random Series')

plt.subplot(1, 2, 2)
plt.scatter(x=df['rand_sq'], y=df['rand'], color='green')
plt.ylabel('Squared Values')
plt.title('Squared Series')

plt.tight_layout()
plt.show()
```

```python
x = np.random.normal(10, 5, 1000)

plt.hist(x)
plt.show()
```

```python
plt.hist(x, bins=40, color='red')
plt.title('Default Bin Placement Demo')
plt.xlabel('Random Values')
plt.show()
```

```python
y = np.random.normal(15, 5, 10000)

plt.hist(x, color ='blue', bins=np.arange(-10, 40), alpha=.5)
plt.hist(y, color='red', bins = np.arange(-10,40), alpha=.5)
plt.title('Manually setting bin placement')
plt.xlabel('Random Values')

plt.show()
```

```python
plt.hist(x, density=True, bins=np.arange(-10,40), alpha=.5)
plt.hist(y, density=True, bins=np.arange(-10,40), alpha=.5)

plt.show()
```
